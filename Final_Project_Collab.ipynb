##Part 1: Age Distribution of People with Heart Disease

```python
import pandas as pd # import data manipulation and analysis library

# initialize the data set
data_url = 'https://raw.githubusercontent.com/csbfx/cs22a/main/heart.csv' # define hyperlink to data
age_heart_df = pd.read_csv(data_url) # import heart disease data as a dataframe
display(age_heart_df) # display heart disease data as a table
```

###B: Create a Dataframe Containing Ages of those with Heart Disease

```python
# create a data subset that contains the ages of all people who have heart disease
age_df = pd.DataFrame(age_heart_df["age"][age_heart_df["heart_disease"]==1]) # extract ages of all those who have heart disease, i.e. heart_disease=1
age_df.columns = ['Age'] # assign a different name to the age column
display(age_df)
```

###C: Box Plot of Age Distribution for those with Heart Disease

```python
import seaborn as age_sns # import data visualization library
import matplotlib.pyplot as age_plt # import plotting library

# create, configure, and display a boxplot of the data
my_color = ["magenta"] # define palette color
age_sns.set_palette(my_color) # set color
age_sns.boxplot(y="Age", data=age_df).set_title("Age") # define data for boxplot
age_plt.show() # draw a boxplot
```

###D: Sample Size of People with Heart Disease

```python
age_sample_size = len(age_df) # calculate total population with heart disease
print("\nThe sample size of people with heart disease is:", age_sample_size)
```

###E: Display Histogram and Probability Density Function of Age

```python
# display a histogram and probability density function of ages
# suppress text output
output = age_sns.histplot(age_df, x='Age', kde=True).set_title(f"Age Distribution of Those Who Have Heart Disease")
```

###F: Calculate Central Tendency of Age Distribution

```python
import statistics # import statistics library

# calculate central tendency
age_mean = sum(age_df['Age'])/age_sample_size # calculate mean
print("\nThe mean age of the population with heart disease is:", "%.1f"%age_mean)

age_median = statistics.median(age_df["Age"]) # calculate median
print("\nThe median age of the population with heart disease is:", age_median)

age_mode = statistics.mode(age_df["Age"]) # calculate mode
print("\nThe mode age of the population with heart disease is:", age_mode)

# calculate standard deviation and variance
age_variance = sum((age_df.Age - age_mean)**2) / age_sample_size # calculate variance
print("\nThe age variance of the population with heart disease is:", round(age_variance,2))

age_SD = statistics.stdev(age_df["Age"]) # calculate standard deviation
print("\nThe age standard deviation of the population with heart disease is:", round(age_SD,2), "\n")
```

###G: Display Normalized Histogram and Probability Density Function of Age

```python
import numpy as age_np # import numerical data library

# creating a normal distribution data
age_norm_df = age_np.random.normal(age_mean, age_SD, 10000)

# display a normalized histogram and probability density function of ages
# suppress text output
output = age_sns.histplot(age_norm_df, kde=True).set_title(f"Normalized Age Distribution of Those Who Have Heart Disease")
```

####4. Calculate the population variance $\sigma^2$   and population standard deviation  𝜎

```python
pop_size = len(cholesterol_df) # calculate total population with heart disease
print("The total population with heart disease is:", pop_size, "out of 303 samples")

# calculating the central tendency
cholesterol_mean = round(cholesterol_df["cholesterol"].mean(),2)
print("The mean cholesterol level of the population with heart disease is:", cholesterol_mean)

cholesterol_median = cholesterol_df["cholesterol"].median()
print("The median cholesterol level of the population with heart disease is:", cholesterol_median)

cholesterol_mode = statistics.mode(cholesterol_df["cholesterol"])
print("The mode cholesterol of the population with heart disease is:", cholesterol_mode)

# calculate standard deviation and variance
cholesterol_variance = sum((cholesterol_df.cholesterol - cholesterol_mean)**2) / pop_size
print("The cholesterol level variance of the population with heart disease is:", round(cholesterol_variance,2))

cholesterol_SD = cholesterol_df["cholesterol"].std()
print("Standard deviation of cholesterol levels for individuals with heart disease:", round(cholesterol_SD,2))
```

####5.Finding the outlier cholesterol levels in those who have heart disease

```python
import pandas as pd

def outliers(cholesterol_df, cholesterol):
    Q1 = cholesterol_df[cholesterol].quantile(0.25)
    Q3 = cholesterol_df[cholesterol].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = cholesterol_df[(cholesterol_df[cholesterol] < lower_bound) | (cholesterol_df[cholesterol] > upper_bound)]
    return outliers

outliers = outliers(cholesterol_df, 'cholesterol')
print(outliers)
```

#Section III - Results & Null Hypotheses Testing

##Part 1: Null Hypothesis Testing for Age Data

### Test 1: Ho for Continuous Age and Heart Disease Data

####1. Define Null Hypothesis

```python
Ho_mean_age = 52
print(f'Null hypothesis is: The mean age of people with heart disease is {Ho_mean_age}.\n')
print(f'Ho: µ = {Ho_mean_age}\n')
print(f'Alternative hypothesis is: The mean age of people with heart disease is not {Ho_mean_age}.\n')
print(f'Ha: µ ≠ {Ho_mean_age} (µ < {Ho_mean_age} or µ > {Ho_mean_age}) Thus, a 2-tailed test will be used\n')
```

####2. Calculate the t-Score

```python
from IPython.display import display, Math
import numpy as age_np # import numerical data library

print(f't-Test formula:\n')
display(Math(r't=\frac{{x̄-μ}}{{s/\sqrt{n}}}'))
print(f'\n where,\n')
print(f'x̄ = {age_mean:.1f} average age for the sample\n')
print(f'μ = {Ho_mean_age} hypothesized mean of the population\n')
print(f's = {age_SD:.2f} standard deviation of the sample\n')
print(f'n = {age_sample_size} sample size for age\n')
age_t_score = (age_mean - Ho_mean_age) / (age_SD / age_np.sqrt(age_sample_size))
print(f't-Score = {age_t_score}')
```

####3. Interpret the t-Score

```python
from scipy import stats # import scientific library

age_pval = stats.t.sf(age_np.abs(age_t_score), age_sample_size-1)*2
print(f'p-Value = {age_pval}')

# check whether we should accept or reject the null hypothesis
# we are using a two-tailed test, with a significance level α=0.05
age_alpha = 0.05
if (age_pval > age_alpha/2) and (age_pval < (1 - age_alpha/2)):
  print(f'{age_alpha/2} < p-Value < {1- age_alpha/2}')
  print(f'Ho Hypothesis is Accepted')
else:
  print(f'Ho Hypothesis is Rejected')
```

###Results from test 1:
By using a z-test that was a 2-tailed test, we were able to find that the mean age of people with heart disease is 52. This is because our null hypothesis failed to be rejected. Our null hypothesis was The mean age of people with heart disease is 52. While our alternative hypothesis is: The mean age of people with heart disease is not 52. When calculating the t-score and finding the p-value, we got a value that was within 0.025 < p-Value < 0.975, meaning that our null was accepted. Moreover, this shows us that the age of people with heart disease is 52.


### Test 2: Ho for Categorical Age and Heart Disease Data

####1. Define Null Hypothesis

```python
Ho_prop_age = 0.40 # Hypothesized proportion
print(f'Null hypothesis is: At least {Ho_prop_age*100}% of people who are 60 and older have heart disease.\n')
print(f'Ho: π >= {Ho_prop_age}\n')
print(f'Alternative hypothesis is: Less than {Ho_prop_age*100}% of people who are 60 and older have heart disease.\n')
print(f'Ha: π < {Ho_prop_age} Thus, a 1-tailed left-sided test will be used\n')
```

####2. Calculate Sample Proportion

```python
# create a dataframe that contains all people who are 60 and older
age_60_df = pd.DataFrame(age_heart_df["heart_disease"][age_heart_df["age"]>=60]) # extract all people who are 60 and older
age_60_df.columns =['HD Yes/No'] # assign a different name to the heart disease status column
display(age_60_df)
```

```python
# Calculate total sample size for people who are 60 and older
age60_sample_size = len(age_60_df)
print("\nThe size of the sample with people 60 years and older is:", age60_sample_size)
```

```python
# Calculate proportion of those who have heart dissease among people who are 60 and older
p_60_heart = age_60_df['HD Yes/No'].sum()/age60_sample_size
print("\nThe proportion of people of 60 and older who have heart disease is:", round(p_60_heart,4))
```

####3. Calculate the z-Score

```python
print(f'z-Test formula:\n')
display(Math(r'z=\frac{{p-π}}{\sqrt{π(1-π)/{N}}}'))
print(f'\n where,\n')
print(f'p = {p_60_heart:.4f} proportion of people 60 and older who have heart disease in the sample\n')
print(f'π = {Ho_prop_age} null hypothesis proportion of the population\n')
print(f'N = {age60_sample_size} sample size for age 60 and older\n')
age_z_score = (p_60_heart - Ho_prop_age) / age_np.sqrt(Ho_prop_age * (1 - Ho_prop_age) / age60_sample_size)
print(f'z-Score = {age_z_score}')
```

####4. Interpret the z-Score

```python
from scipy import stats # import scientific library

age_prop_pval = stats.norm.sf(age_np.abs(age_z_score))
print(f'p-Value = {age_prop_pval}')

# check whether we should accept or reject the null hypothesis
# we are using a one-tailed test, with a significance level α=0.05
age_alpha = 0.05
if (age_pval > age_alpha):
  print(f'p-Value > {age_alpha}')
  print(f'Ho Hypothesis is Accepted')
else:
  print(f'Ho Hypothesis is Rejected')
```

###Results from Test 2:
####By using a one-tailed t-test, we were able to find that at least 40% of people who are 60 and older have developed heart disease. This is because our null hypothesis failed to be rejected. Our null hypothesis was that at least 40.0% of people who are 60 and older have heart disease. While our alternative hypothesis is that less than 40.0% of people who are 60 and older have heart. When calculating the z-score and finding the p-value, we got a value that was greater than 0.05, which was our alpha or critical value. Meaning that our null was accepted.


## Part 2 Null Hypothesis Testing For Cholesterol Data

###Test 3:

```python
Ho_mean_ch = 234
print(f'Null hypothesis is: The mean cholesterol of people with heart disease is {Ho_mean_ch}.\n')
print(f'Ho: µ = {Ho_mean_ch}\n')
print(f'Alternative hypothesis is: The mean cholesterol of people with heart disease is not {Ho_mean_ch}.\n')
print(f'Ha: µ ≠ {Ho_mean_age} (µ < {Ho_mean_ch} or µ > {Ho_mean_ch}) Thus, a 2-tailed test will be used\n')
```

```python
from IPython.display import display, Math
import numpy as age_np # import numerical data library

print(f't-Test formula:\n')
display(Math(r't=\frac{{x̄-μ}}{{s/\sqrt{n}}}'))
print(f'\n where,\n')
print(f'x̄ = {cholesterol_mean:.1f} average cholesterol for the sample\n')
print(f'μ = {Ho_mean_ch} hypothesized mean of the population\n')
print(f's = {cholesterol_SD:.2f} standard deviation of the sample\n')
print(f'n = {pop_size} sample size for cholesterol\n')
ch_t_score = (cholesterol_mean - Ho_mean_ch) / (cholesterol_SD / np.sqrt(pop_size))
print(f't-Score = {ch_t_score}')
```

```python
from scipy import stats # import scientific library

ch_pval = stats.t.sf(np.abs(ch_t_score), pop_size-1)*2
print(f'p-Value = {ch_pval}')

# check whether we should accept or reject the null hypothesis
# we are using a two-tailed test, with a significance level α=0.05
ch_alpha = 0.05
if (ch_pval > ch_alpha/2) and (ch_pval < (1 - ch_alpha/2)):
  print(f'{ch_alpha/2} < p-Value < {1- ch_alpha/2}')
  print(f'Ho Hypothesis is Accepted')
else:
  print(f'Ho Hypothesis is Rejected')
```

###Results Test 3:
####By using a two-tailed t-test, we were able to find that the mean cholesterol level of people with heart disease is 234 mg/dL. This is because our null hypothesis failed to be rejected. Our null hypothesis was The mean cholesterol of people with heart disease is 234 mg/dL, while our alternative hypothesis was that the mean cholesterol of people with heart disease is not 234 mg/dL. When calculating the z-score and finding the p-value, we got a p-value that was in between 0.025 < p-Value < 0.975, meaning that our null hypothesis was accepted.
